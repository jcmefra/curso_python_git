# Visualización de datos

## Gradiente Geotérmico

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_gradient = pd.read_csv('datos/data_weights.csv', sep = ',')
```

```python
df_gradient.head()
```

```python
df_gradient.columns
```

```python
percentil_95 = df_gradient['Apparent Geothermal Gradient (°C/Km)'].quantile(0.95)
```

```python
percentil_50 = df_gradient['Apparent Geothermal Gradient (°C/Km)'].quantile(0.50)
```

```python
percentil_1 = df_gradient['Apparent Geothermal Gradient (°C/Km)'].quantile(0.1)
```

```python
media = df_gradient['Apparent Geothermal Gradient (°C/Km)'].mean()
```

```python
sns.histplot(df_gradient['Apparent Geothermal Gradient (°C/Km)'], bins=60)
plt.axvline(percentil_95, color='red', linestyle='--')
plt.axvline(percentil_50, color='blue', linestyle='--')
plt.axvline(percentil_1, color='green', linestyle='--')
plt.axvline(media, color='yellow', linestyle='--')
plt.show()
```

```python
sns.kdeplot(df_gradient['Apparent Geothermal Gradient (°C/Km)'], shade=True)
```

```python
df_gradient.columns
```

```python
selected_columns = ['Apparent Geothermal Gradient (°C/Km)', 
       'Elevation (m)', 'Surface Temperature (°C)',
       'Moho Depth (m)', 'Curie Depth (Km)', 'Nearest Basement', 
       'Magnetic Anomaly (nT)', 'Free Air Anomaly (mGal)',
       'Bouguer Anomaly (mGal)']
```

```python
selected_data = df_gradient[selected_columns]
```

```python
corr_matrix = selected_data.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
